```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('datasets/advertising.csv')
df
```

```python
df.info()
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
df.info()
```

```python
def ext_season(x):
    if x.month in [12,1,2]:
        return 'Winter'
    elif x.month in [3,4,5]:
        return 'Spring'
    elif x.month in [6,7,8]:
        return 'Summer'
    else:
        return 'Autumn'

def ext_period(x):
    if x.hour in range(0,13):
        return 'Morning'
    elif x.hour in range(13,19):
        return 'AfterNoon'
    else:
        return 'Evening'
```

```python
df['Season']=df['Timestamp'].apply(ext_season)
df['Period']=df['Timestamp'].apply(ext_period)
df
```

```python
df=pd.get_dummies(df,columns=['City','Country','Season','Period'],drop_first=True)
df
```

```python
x=df.drop(['Clicked on Ad','Timestamp','Ad Topic Line'],axis=1)
y=df['Clicked on Ad']
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
y_train.value_counts()
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
#x_train[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]=scaler.fit_transform(x_train[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']])
#x_test[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]=scaler.transform(x_test[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']])
#or
#x[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]=scaler.fit_transform(x[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']])
#splitting
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score,fbeta_score,classification_report
```

```python
models={
    'LR':LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'DT':DecisionTreeClassifier(),
    'SVC':SVC(),
    'NB':GaussianNB(),
    'XGC':XGBClassifier(),
    'RF':RandomForestClassifier()
    
}
```

```python
for name,model in  models.items():
    print(f'using {name}: ')
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f'Training Accuracy :{accuracy_score(y_train,model.predict(x_train))}')
    print(f'Testing Accuracy :{accuracy_score(y_test,y_pred)}')
    print(f'Confusion matrix:\n {confusion_matrix(y_test,y_pred)}')
    print(f'Recall: {recall_score(y_test,y_pred)}')
    print(f'precision: {precision_score(y_test,y_pred)}')
    print(f'F1-score: {f1_score(y_test,y_pred)}')
    print(f'Fbeta-score: {fbeta_score(y_test,y_pred,beta=0.5)}')
    print(classification_report(y_test,y_pred))
    print('-'*33)
    
```

```python
df=pd.read_csv('datasets/sentiment.csv',sep='\t',names=['liked','tweet'])
df
```

```python
x=df['tweet']
y=df['liked']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words='english')
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
x_train
```

```python
tf_df=pd.DataFrame(x_train.toarray(),columns=vectorizer.get_feature_names())
tf_df
```

```python
from sklearn.naive_bayes import MultinomialNB
models={
    'LR':LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'DT':DecisionTreeClassifier(),
    'SVC':SVC(),
    'NB':MultinomialNB(),
    'XGC':XGBClassifier(),
    'RF':RandomForestClassifier()
    
}
```

```python


for name,model in  models.items():
    print(f'using {name}: ')
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f'Training Accuracy :{accuracy_score(y_train,model.predict(x_train))}')
    print(f'Testing Accuracy :{accuracy_score(y_test,y_pred)}')
    print(f'Confusion matrix:\n {confusion_matrix(y_test,y_pred)}')
    print(f'Recall: {recall_score(y_test,y_pred)}')
    print(f'precision: {precision_score(y_test,y_pred)}')
    print(f'F1-score: {f1_score(y_test,y_pred)}')
    print(f'Fbeta-score: {fbeta_score(y_test,y_pred,beta=0.5)}')
    print(classification_report(y_test,y_pred))
    print('-'*33)
    
```
