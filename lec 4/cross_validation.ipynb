```python
import pandas as pd
df=pd.read_csv('Model_Selection/AdultIncome.csv')
df
```

```python
df.info()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
x=df.drop('IncomeClass_ >50K',axis=1)
x
```

```python
y=df['IncomeClass_ >50K']
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x[['age','hours per week']]=scaler.fit_transform(x[['age','hours per week']])
x
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict,cross_val_score,cross_validate
```

```python
model=DecisionTreeClassifier(random_state=42)
CV_result=cross_validate(model,x,y,cv=10,n_jobs=-1,return_train_score=True)
CV_result
```

```python
CV_result['train_score'].mean()
```

```python
CV_result['test_score'].mean()
```

```python
df=pd.read_csv('Model_Selection/Advertising.csv')
df
```

```python
df.info()
```

```python
from datetime import datetime
df['Timestamp']=pd.to_datetime(df['Timestamp'],format='%Y-%m-%d %H:%M:%S')
df
```

```python
df.info()
```

```python
df=pd.get_dummies(df,columns=['City','Country'],drop_first=True)
df
```

```python
def ext_pod(x):
    if x.hour in range(0,12):
        return 'Morning'
    elif x.hour in range(12,19):
        return 'AfterNoon'
    else:
        return 'Evening'
df['Period']=df['Timestamp'].apply(ext_pod)
df
```

```python
def ext_season(x):
    if x.month in [12,1,2]:
        return 'winter'
    elif x.month in [3,4,5]:
        return 'Spring'
    elif x.month in [6,7,8]:
        return 'Summer'
    else:
        return 'Autumn'
df['Season']=df['Timestamp'].apply(ext_season)
df
```

```python
df=pd.get_dummies(df,columns=['Period','Season'],drop_first=True)
df
```

```python
x=df.drop(['Ad Topic Line','Timestamp','Clicked on Ad'],axis=1)
y=df['Clicked on Ad']
```

```python
CV_result=cross_validate(RandomForestClassifier(),x,y,n_jobs=-1,cv=10,return_train_score=True)
CV_result
```

```python
CV_result['test_score'].mean()
```
